
* Numeric data types, represents numeric value.
* A Numeric value can be an integer, a floating number, and complex number.

#### Note: We can use type() function to determine the type.

## 1. Integers
* All positive & negative whole numbers without fractions/decimals.
* The value is represented by 'int' class.

```python
enrollment_number = 17045
print("enrollment number of a student =",enrollment_number)
print("It's data type =",type(enrollment_number))
```

```python
#add two numbers
a = 5
b = 4
c = a+b
print("a + b =",c)
print("Data type of c = ",type(c))
```

## 2. float
* All the real numbers with a floating-point representation.
* The value is represented by 'float' class.

```python
tata_motors_quantity = 101
price = 960.60

stock_portfolio_size = (tata_motors_quantity * price)
print("stock_portfolio_size = ",stock_portfolio_size)
print("Data type of stock_portfolio_size = ",type(stock_portfolio_size))
```

## 3.Complex numbers
* complex number = (real number part) + (imaginary part)
* The value is represented by 'complex' class.

```python
var = 1+2j
print("var =",var)
print("Data type of var =",type(var))
```
